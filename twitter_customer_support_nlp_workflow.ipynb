**Text Cleaning Steps:**

1. **Convert all text to lowercase** – This helps ensure consistency, so that "Great" and "great" are treated the same.
2. **Remove punctuation** – Getting rid of symbols like commas, periods, and question marks reduces noise in the data.
3. **Remove stopwords** – Common words like “and,” “the,” or “is” that don't add much meaning are filtered out.
4. **Remove very frequent words** – Words that appear too often across all text (but carry little specific meaning) can be excluded.
5. **Remove very rare words** – Similarly, extremely uncommon words may not contribute meaningful patterns and can be dropped.
6. **Apply stemming** – This reduces words to their root form by chopping off suffixes (e.g., “running” becomes “run”), though it may be rough.
7. **Apply lemmatization** – A more refined alternative to stemming, this reduces words to their base form using vocabulary and grammar (e.g., “better” becomes “good”).
8. **Keep emojis** – Emojis are retained since they carry strong emotional and sentiment value, which is important for sentiment analysis.
9. **Remove emoticons** – Emoticons like `:)` or `:(` can be cleaned out or handled separately depending on use case.
10. **Convert emoticons to words** – Before removal, translate emoticons into their textual meaning (e.g., `:)` becomes "happy face").
11. **Convert emojis to words** – Emojis are translated to words (e.g., 😂 becomes "laughing face") to preserve their sentiment meaning.
12. **Remove URLs** – Links are not usually helpful for text meaning and are stripped out.
13. **Remove HTML tags** – Clean out leftover tags from scraped content like `<div>` or `<br>`.
14. **Expand chat words** – Convert shorthand/slang into full words (e.g., “u” → “you”, “brb” → “be right back”).
15. **Correct spelling** – Fix common typos or misspellings to improve accuracy in tokenization and model understanding.



# Lowering case

Lowercasing is a basic but important text preprocessing step. It involves converting all characters in the text to lowercase so that variations like "text", "Text", and "TEXT" are recognized as the same word, ensuring consistency during analysis.

Most modern vectorizers and tokenizers, such as TfidfVectorizer in scikit-learn and Tokenizer in Keras, automatically convert all text to lowercase by default. However, depending on the use case, this behavior can be disabled.

In our case, capital letters might indicate strong emotion—like anger or urgency—so preserving the original casing could help capture valuable sentiment signals in the text.

```python
df["text_lower"] = df["text"].str.lower()
df.head()
```

# Removal of Punctuations



> In Python, `string.punctuation` provides a default set of common punctuation characters, which includes:
> \`!"#\$%&'()\*+,-./:;<=>?@\[]^\_{|}\~\`\`
> This set can be customized based on your needs—you can easily add more symbols to it or remove specific ones, depending on the context of your text preprocessing.


```python
import string
PUNCT_TO_REMOVE = string.punctuation
print(PUNCT_TO_REMOVE)
```

```python
def remove_punctuation(text):
    """custom function to remove the punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

df["text_to_punct"] = df["text"].apply(lambda text: remove_punctuation(text))
df.head()
```

 # Removal of stopwords

Stopwords are common words in a language, such as “the,” “a,” and “and,” that usually don’t carry significant meaning on their own and are often removed during text processing.

```python
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

df["text_to_stop"] = df["text_to_punct"].apply(lambda text: remove_stopwords(text))
df.head()
```

# Removal of Frequent words


This step involves removing words that appear very frequently across the entire corpus, as they often add little meaningful information. When using methods like TF-IDF, this effect is handled automatically since TF-IDF downweights common words. However, if not using TF-IDF, we first identify these frequent words manually and then remove them to improve the quality of our text data.

Improved Analysis: Removing frequent words helps the text focus on unique and more meaningful content, which can enhance tasks like sentiment analysis and topic modeling.

Reduced Dimensionality: Eliminating these common words can also lower the data’s dimensionality, making it simpler and faster for machine learning models to process.

Customization: This approach lets you tailor the list of frequent words to your specific dataset and goals. By adjusting the threshold—such as removing the top 10 most frequent words—you control which terms get filtered out.

```python
from collections import Counter
cnt = Counter() #returns a list of tuples containing the most common words and their respective word counts in descending order.
for text in df["text_to_stop"].values:
    for word in text.split():
        cnt[word] += 1

cnt.most_common(10)
```

```python
FREQWORDS = set() #This line creates a set called FREQWORDS that contains the most common 10 words in the corpus.

most_common = cnt.most_common(10)
for w, wc in most_common: #The list comprehension [w for (w, wc) in cnt.most_common(10)] extracts only the words (not their counts) from the list of tuples.
    FREQWORDS.add(w)


def remove_freqwords(text):
    """Custom function to remove the frequent words"""

    # Split the input text into words
    words = str(text).split()

    # Initialize an empty list to store words that are not in FREQWORDS
    filtered_words = []

    # Iterate through the words in the input text
    for word in words:
        # Check if the word is not in FREQWORDS
        if word not in FREQWORDS:
            # If it's not in FREQWORDS, add it to the filtered_words list
            filtered_words.append(word)

    # Join the filtered words into a single string with spaces in between
    cleaned_text = " ".join(filtered_words)
    #This line is the implementation of the remove_freqwords function. It takes the input text, splits it into words, and then uses a list comprehension to include only the words that are not in the FREQWORDS set. Finally, it joins these words back into a string with spaces in between, effectively removing the frequent words.

    # Return the cleaned text
    return cleaned_text


df["text_to_stopfreq"] = df["text_to_stop"].apply(lambda text: remove_freqwords(text))
#This line creates a new column in the DataFrame df named "text_wo_stopfreq". It applies the remove_freqwords function to each element in the "text_wo_stop" column of the DataFrame using the apply() method.
df.head()
```

# Removal of Rare words


Removal of Rare Words: Rare words are those that appear very infrequently in a dataset, often just once or twice. While some rare words may carry important information, many are typos, noise, or overly specific terms that don’t contribute meaningfully to the overall analysis. Removing these rare words helps reduce noise and complexity in the data, improving model performance and generalization. It also reduces the size of the vocabulary, making processing more efficient. Like frequent words, the threshold for what counts as “rare” can be adjusted based on the dataset and task requirements.

```python


n_rare_words = 10 #threshold can be changed

# Get the word frequency counts in descending order
most_common_words = cnt.most_common()

# Initialize an empty set to store rare words
RAREWORDS = set()

# Calculate the index for selecting the least common words
index_to_select = len(most_common_words) - n_rare_words

# Iterate through the most common words in descending order
for i, (word, wc) in enumerate(most_common_words):
    # Check if the current word should be considered rare
    if i >= index_to_select:
        RAREWORDS.add(word)


def remove_rarewords(text):
    """Custom function to remove the rare words"""

    # Split the input text into words
    words = str(text).split()

    # Initialize an empty list to store words that are not in RAREWORDS
    filtered_words = []

    # Iterate through the words in the input text
    for word in words:
        # Check if the word is not in RAREWORDS
        if word not in RAREWORDS:
            # If it's not in RAREWORDS, add it to the filtered_words list
            filtered_words.append(word)

    # Join the filtered words into a single string with spaces in between
    cleaned_text = " ".join(filtered_words)

    # Return the cleaned text
    return cleaned_text




df["text_to_stopfreqrare"] = df["text_to_stopfreq"].apply(lambda text: remove_rarewords(text))
df.head()
```

# Stemming

Stemming is the technique of shortening words to their base or root form by removing suffixes or endings. For instance, words like “walks” and “walking” would both be reduced to “walk” through stemming.

```python
# Drop the two columns
df.drop(["text_to_stopfreq", "text_to_stopfreqrare"], axis=1, inplace=True)
df
```

```python
from nltk.stem.porter import PorterStemmer
# Import the Porter Stemmer
from nltk.stem import PorterStemmer

# Initialize the Porter Stemmer
stemmer = PorterStemmer()

def stem_words(text):
    """Custom function to stem words in text"""

    # Split the input text into words
    words = text.split()

    # Initialize an empty list to store stemmed words
    stemmed_words = []

    # Iterate through the words in the input text
    for word in words:
        # Apply stemming to the word and add the stemmed word to the list
        stemmed_word = stemmer.stem(word)
        stemmed_words.append(stemmed_word)

    # Join the stemmed words into a single string with spaces in between
    stemmed_text = " ".join(stemmed_words)

    # Return the stemmed text
    return stemmed_text



df["text_stemmed"] = df["text"].apply(lambda text: stem_words(text))
df.head()
#We can see that words like private and propose have their e at the end chopped off due to stemming. This is not intented. What can we do fort hat? We can use Lemmatization in such cases.
```

# Lemmatization

Lemmatization, like stemming, reduces inflected words to their base form, but it goes a step further by ensuring that the root word (called the lemma) is a valid word in the language. Because of this, lemmatization is generally slower than stemming. Depending on your need for speed versus accuracy, you can choose to use either stemming or lemmatization.

For example, we can use the `WordNetLemmatizer` from the `nltk` library to perform lemmatization on our text.


```python
df
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
# Import the WordNet Lemmatizer
from nltk.stem import WordNetLemmatizer

# Initialize the WordNet Lemmatizer
lemmatizer = WordNetLemmatizer()

def lemmatize_words(text):
    """Custom function to lemmatize words in text"""

    # Split the input text into words
    words = text.split()

    # Initialize an empty list to store lemmatized words
    lemmatized_words = []

    # Iterate through the words in the input text
    for word in words:
        # Apply lemmatization to the word and add the lemmatized word to the list
        lemmatized_word = lemmatizer.lemmatize(word)
        lemmatized_words.append(lemmatized_word)

    # Join the lemmatized words into a single string with spaces in between
    lemmatized_text = " ".join(lemmatized_words)

    # Return the lemmatized text
    return lemmatized_text


df["text_lemmatized"] = df["text"].apply(lambda text: lemmatize_words(text))
df.head()
```

# Removing Emojis and Emoticons

**Removing Emojis and Emoticons:** Emojis and emoticons are visual elements used in text to express emotions or reactions. While emojis are standardized Unicode characters (like 😊 or 😂), emoticons are combinations of keyboard symbols (like :-) or :-(). Depending on your analysis goals, you may want to remove them to reduce noise, especially if your model doesn’t process them effectively. However, since both carry important sentiment information, in some cases, it’s better to convert them into descriptive words rather than simply deleting them. This preserves their emotional meaning while keeping the text clean.


```python
# Reference : https://gist.github.com/slowkow/7a7f61f495e3dbb7e3d767f97bd7304b
def remove_emoji(string):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)

remove_emoji("game is on 🔥🔥")
```

```python
import re

def remove_emoticons(text):
    """Custom function to remove emoticons from text"""

    # Create a regular expression pattern for emoticons
    emoticon_pattern = re.compile(u'(' + u'|'.join(k for k in EMOTICONS) + u')')

    # Use the re.sub() method to replace emoticons with an empty string
    cleaned_text = emoticon_pattern.sub(r'', text)

    # Return the cleaned text
    return cleaned_text


remove_emoticons("Hello :-)")
```

# Conversion of Emoticon to Words

```python
import re

def convert_emoticons(text):
    """Custom function to convert emoticons to textual representations"""

    # Iterate through each emoticon in the EMOTICONS dictionary
    for emot in EMOTICONS:
        # Replace the emoticon in the text with its textual representation
        replacement = "_".join(EMOTICONS[emot].replace(",", "").split())
        text = re.sub(u'(' + emot + u')', replacement, text)

    # Return the text with converted emoticons
    return text


text = "Hello :-) :-)"
convert_emoticons(text)
```

# Remove of URLS

Removal of URLs: URLs often appear in text but usually don’t contribute meaningful information for analysis. They can add noise and unnecessary complexity, so it’s common to remove them during preprocessing to keep the focus on the actual content. This helps improve the quality of the data and the performance of downstream models.

```python
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)
```

# Chat Words Conversion

Chat Words Conversion: In informal text, especially on social media and messaging platforms, people often use slang, abbreviations, or shorthand like “brb” (be right back) or “u” (you). Converting these chat words into their full, standard forms helps improve the clarity and consistency of the text, making it easier for models to understand and analyze the content accurately.

```python
chat_words_map_dict = {}
chat_words_list = []
for line in chat_words_str.split("\n"):
    if line != "":
        cw = line.split("=")[0]
        cw_expanded = line.split("=")[1]
        chat_words_list.append(cw)
        chat_words_map_dict[cw] = cw_expanded
chat_words_list = set(chat_words_list)

def chat_words_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_words_list:
            new_text.append(chat_words_map_dict[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)

chat_words_conversion("one minute BRB")
```

# Spelling Correction

Spelling Correction: Text data often contains typos, misspellings, or informal variations of words. Correcting these errors helps standardize the text, reducing noise and improving the accuracy of language models. By fixing spelling mistakes, we ensure that similar words are recognized as the same, which enhances overall text analysis and model performance.

```python
!pip install pyspellchecker
```

```python
from spellchecker import SpellChecker

spell = SpellChecker()
def correct_spellings(text):
    corrected_text = []
    misspelled_words = spell.unknown(text.split())


    for word in text.split():
        if word in misspelled_words:
            corrected_text.append(spell.correction(word))
        else:
            corrected_text.append(word)
    return " ".join(corrected_text)

text = "speling correctin"
correct_spellings(text)
```
